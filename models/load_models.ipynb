```python
# Example text for testing
text = (
    "De Autoriteit Consument en Markt besluit "
    "op grond van artikel 2a, eerste lid, "
    "onderdeel a, van de Gaswet een ontheffing "
    "te verlenen aan Schiphol Nederland B.V. "
)
```

## NER model

```python
from src.load_model import load_ner_model, label_map
from src.inference import predict_ner_entities

model_ner, tokenizer_ner = load_ner_model(
    "NER-silver-to-gold" # or "NER-gold", "NER-silver"
)

preds = predict_ner_entities(
    text,
    model_ner,
    tokenizer_ner,
    label_map=label_map
)

for p in preds:
    print(
        f"{p['label']:23} | "
        f"{p['text']}"
    )
```

```python
from src.load_model import load_nerre_model
from src.inference import predict_entities

model, tokenizer, cfg = load_nerre_model(
    "NERRE-silver-to-gold" # or "NERRE-gold", "NERRE-silver"
)

preds = predict_entities(
    text,
    model,
    tokenizer,
    cfg
)

for p in preds:
    print(
        f"{p['label']:23} | "
        f"{p['text']}"
    )
```
